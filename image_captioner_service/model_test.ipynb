```python
from huggingface_hub import snapshot_download

local_dir = "models/Qwen2.5-VL"

snapshot_download(
    repo_id   = "Qwen/Qwen2.5-VL-7B-Instruct",
    local_dir = local_dir,       # your target folder
    # resume    = True
)
```

```python
from transformers import AutoProcessor, AutoModelForVision2Seq

local_dir = "./llm_models/Qwen2.5-VL"

processor = AutoProcessor.from_pretrained(local_dir, local_files_only=True)
model     = AutoModelForVision2Seq.from_pretrained(local_dir, local_files_only=True)
```

```python
from PIL import Image
async def caption_image(file: str):
    image = Image.open(file.file).convert("RGB")
    inputs = processor(images=image, return_tensors="pt")
    outputs = model.generate(**inputs)
    caption = processor.decode(outputs[0], skip_special_tokens=True)
    return {"caption": caption}
```

```python
caption_image(file="./sample-files/MBS.png")
```
